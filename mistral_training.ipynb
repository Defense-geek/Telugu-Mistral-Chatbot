```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

# 4bit pre quantized models we support for 4x faster downloading + no OOMs.
fourbit_models = [
    "unsloth/mistral-7b-bnb-4bit",
    "unsloth/mistral-7b-instruct-v0.2-bnb-4bit",
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/mistral-7b-v0.3", # Choose ANY! eg teknium/OpenHermes-2.5-Mistral-7B
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

```python
from datasets import load_dataset

# Load dataset
dataset = load_dataset("json", data_files="mistral_training_data.jsonl", split="train")

# Tokenization function
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=2048)

# Tokenize dataset
tokenized_datasets = dataset.map(tokenize_function, batched=True)
print("✅ Tokenization successful!")
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./mistral_finetuned",
    per_device_train_batch_size=1,  # Reduce to avoid OOM issues
    gradient_accumulation_steps=8,  # Helps simulate a larger batch
    num_train_epochs=1,  # Start with 1 epoch, increase later
    save_steps=500,  # Save less frequently
    logging_dir="./logs",
    logging_steps=50,
    eval_strategy="no",  # Use "steps" if you add validation data
    learning_rate=1e-4,
    weight_decay=0.01,
    fp16=True,  # Mixed precision training for better efficiency
    push_to_hub=False,
)
```

```python
import torch
from transformers import Trainer, TrainingArguments

# Updated Custom Trainer class compatible with Unsloth
class CustomTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        # Support for num_items_in_batch parameter used by Unsloth
        labels = inputs.get("input_ids").clone()  # Ensure labels exist
        outputs = model(**inputs, labels=labels)  # Compute loss using labels
        loss = outputs.loss  # Extract loss

        return (loss, outputs) if return_outputs else loss

# Training arguments
training_args = TrainingArguments(
    output_dir="./mistral_finetuned",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    num_train_epochs=1,
    logging_dir="./logs",
    logging_steps=10,
    save_strategy="epoch",
    evaluation_strategy="no",  # No evaluation during training
    fp16=True,  # Mixed precision for faster training
)

# Initialize Custom Trainer
trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets,
    tokenizer=tokenizer,
)

# Start Training
trainer.train()
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Path to your saved model
model_path = "/content/drive/MyDrive/mistral_finetuned_complete"  # Use your saved directory
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path)

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

# Set up a simple function to generate text without using model.generate()
def generate_text_simple(prompt, max_tokens=100):
    # Tokenize the prompt
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)

    # Get the initial prompt length to track new tokens
    input_length = input_ids.shape[1]

    # Generate one token at a time
    for _ in range(max_tokens):
        with torch.no_grad():
            outputs = model(input_ids)

        # Get the next token prediction (last token)
        next_token_logits = outputs.logits[:, -1, :]

        # Simple sampling - you can adjust this for different generation styles
        next_token = torch.argmax(next_token_logits, dim=-1).unsqueeze(-1)

        # Append the next token to input_ids
        input_ids = torch.cat([input_ids, next_token], dim=-1)

        # Stop if we generate an EOS token
        if next_token.item() == tokenizer.eos_token_id:
            break

    # Decode the generated text
    generated_text = tokenizer.decode(input_ids[0][input_length:], skip_special_tokens=True)
    full_text = prompt + generated_text

    return full_text

# Test with a prompt
test_prompt = "The best way to learn machine learning is"
try:
    result = generate_text_simple(test_prompt)
    print(f"Prompt: {test_prompt}")
    print(f"Generated: {result}")
except Exception as e:
    print(f"Error: {str(e)}")

# Interactive testing loop
def interactive_testing():
    print("\nEnter prompts to test the model (type 'exit' to quit):")
    while True:
        user_prompt = input("\nYour prompt: ")
        if user_prompt.lower() == "exit":
            break

        try:
            generated = generate_text_simple(user_prompt)
            print(f"\nGenerated output:\n{generated}")
        except Exception as e:
            print(f"Error: {str(e)}")

# Run the interactive testing
interactive_testing()
```
